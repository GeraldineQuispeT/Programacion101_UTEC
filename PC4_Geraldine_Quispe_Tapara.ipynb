## Pregunta 2

1. Crea un DataFrame df_filtrado con todas las filas de df_diabetes donde Glucose > 120.

```python
df_filtrado = df_diabetes[df_diabetes['Glucose']>120]
print(f'La cantidad de filas del df_filtrado es: {df_filtrado.shape[0]}')
```

2. Agrega a df_filtrado una columna BMI_Category que clasifique el BMI así:
* "Bajo" si BMI < 18.5
* "Normal" si 18.5 ≤ BMI < 25
* "Sobrepeso" si 25 ≤ BMI < 30
* "Obeso" si BMI ≥ 30


```python
df_filtrado = df_filtrado.copy()
df_filtrado['BMI_Category'] = None
df_filtrado.loc[df_filtrado['BMI']<18.5,'BMI_Category'] = 'Bajo'
df_filtrado.loc[(df_filtrado['BMI']>=18.5) & (df_filtrado['BMI']<25),'BMI_Category'] = 'Normal'
df_filtrado.loc[(df_filtrado['BMI']>=25) & (df_filtrado['BMI']<30),'BMI_Category'] = 'Sobrepeso'
df_filtrado.loc[df_filtrado['BMI']>=30,'BMI_Category'] = 'Obeso'
```

```python
df_filtrado['BMI_Category'].value_counts()
```

```python
# Existe un gran numero de personas con sobrepeso ya que muestran BMI>30, lo cual es alarmante y esta relacionado a su diabetes y alta glucosa
```

## Pregunta 3

1. Carga age_groups.csv en un DataFrame df_age_groups.

```python
df_age_groups = pd.read_csv('age_groups.csv')
```

2. Haz un merge de df_diabetes con df_age_groups por la columna Age. El resultado se llamará
df_combinado. Debe ser un left merge para mantener todas las filas de df_diabetes.

```python
df_combinado = pd.merge(df_diabetes, df_age_groups, on='Age')
print(f'Las cantidad de filas en df_diabetes es {df_diabetes.shape[0]} y en df_combinado es {df_combinado.shape[0]}')
```

3. Muestra las 5 primeras filas de df_combinado para verificar que la columna AgeGroup se haya
agregado correctamente.

```python
df_combinado.head()
```

## Pregunta 4

1. Agrupa df_combinado por AgeGroup.


```python
agrupado = df_combinado.groupby('AgeGroup')
agrupado
```

2. Calcula el promedio de Glucose y el promedio de BMI para cada grupo de edad.

```python
resultado = agrupado.agg({'Glucose': 'mean', 'BMI': 'mean'})
resultado
```

3. Ordena el resultado por el promedio de Glucose de mayor a menor.

```python
df_resultado_gluc_desc = resultado.sort_values(by='Glucose', ascending=False)
```

4. Muestra el DataFrame resultante.

```python
df_resultado_gluc_desc.head()
```

```python
# Las personas en la categoria de adulto mayor son los que normalmente presentan mayores problemas de glucosa a diferencia de los adultos y jovenes.
```
